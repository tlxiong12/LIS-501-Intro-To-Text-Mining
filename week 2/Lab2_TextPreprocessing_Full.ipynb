## Use NLTK for Text Preprocessing

#### Sentence Segmentation

```python
from nltk.tokenize import sent_tokenize

nltk_sentences = sent_tokenize(text)
print("Number of sentences: ", len(nltk_sentences))
print("\n")

for s in nltk_sentences:
    print (s)
```

#### Tokenization

```python
from nltk.tokenize import word_tokenize

nltk_words = word_tokenize(text)
print(nltk_words)
```

#### Case Normalization

```python
lowercase_text = text.lower()
print(lowercase_text)

lowercase_words = word_tokenize(lowercase_text)
print(lowercase_words)
```

#### Part-of-speech Tagging

```python
from nltk import pos_tag

pos_tags = pos_tag(lowercase_words)
print(pos_tags)
```

#### Remove Stop Words

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
filtered_sentence = [word for word in nltk_words if word not in stop_words]
print(filtered_sentence)
```

#### Stemming

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_sentence]
print(stemmed_words)
```

#### Lemmatization

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in stemmed_words]
print(lemmatized_words)
```

## Use Spacy for Text Preprocessing

```python
doc = nlp(text)

spacy_sentences = list(doc.sents)
print("Number of sentences: ", len(spacy_sentences))
print("\n")

for s in spacy_sentences:
    print (s)
```

#### Tokenization

```python
spacy_tokens = list(doc)
print(spacy_tokens)
```

```python
lowercase_text = text.lower()
print(lowercase_text)

lowercase_doc = nlp(lowercase_text)
lowercase_tokens = list(lowercase_doc)
print(lowercase_tokens)
```

#### Part-of-speech Tagging

```python
for token in lowercase_doc:
    print(f"{token.text} {token.pos_} {token.tag_}")
```

#### Remove Stop Words

```python
filtered_tokens = [token.text for token in lowercase_doc if not token.is_stop and not token.is_punct and not token.is_space]
print(filtered_tokens)
```

#### Stemming

Spacy does not have a built-in Porter stemmer or any other traditional stemming function because it relies exclusively on lemmatization.

#### Lemmatization

```python
for token in lowercase_doc:
    if not token.is_stop and not token.is_punct and not token.is_space:
        print(f"Token: {token.text}, Lemma: {token.lemma_}")
```

```python
from collections import Counter

frequent_nouns = ['NN', 'NNP', 'NNS', 'NNPS']

raw_lower = raw.lower()
tokens = word_tokenize(raw_lower)
tagged = pos_tag(tokens)

token_list = []

for word, tag in tagged:
    if tag in frequent_nouns:
        token_list.append(tag)

Counter(token_list)
```

```python
frequent_nouns = ['NN', 'NNP', 'NNS', 'NNPS']

tokens = [token.text for token in text]   
tagged = pos_tag(tokens)          

noun_count = Counter()

for word, tag in tagged:
    if tag in frequent_nouns:
        noun_count[tag] += 1

noun_count
```
