```python
import os
from collections import Counter
```

```python
import spacy
import en_core_web_sm

nlp = en_core_web_sm.load()
```

```python
def get_corpus(fd_path): 
    '''
       Input: 
            fd_path: a string of input folder path 
            
       Output:
            corpus_to_string: a string of content from all text files in the input folder
    '''
    
    data = []
    files = os.listdir(fd_path)
    
    # read the content of each text file
    for fname in files:
        if fname.endswith('.txt'):
            f_content = open(fd_path + '/' + fname, mode='r', encoding = 'utf-8').read()
            data.append(f_content)
    
    # aggreage the list of texts into a single string
    corpus_to_string = ' '.join(data)
    
    return corpus_to_string
    


def get_frequent_words(corpus, k):
    '''
       Input: 
            corpus: a string of corpus
            k: the number of top frequent words to be selected, here k should be 20
       Output:
            top_k_words: a list of top k most frequent words and their frequency
    '''
    
    # get the list of words in the corpus (no pre-processing)
    # words = corpus.split(' ')

    # get the list of words in the corpus with spacy pre-processing
    doc = nlp(corpus)
    words = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct and not token.is_space]
    
    
    # get top k most frequent words and their frequency
    corpus_stat = Counter(words)
    del corpus_stat['']
    top_k_words = corpus_stat.most_common(k)
    
    return top_k_words


def write_to_file(towrite_fpath, top_words):
    '''
       Input:
            towrite_fname: a string of output file name
            top_words: a list to top frequent words and their frequency to be saved
    '''
    with open(f'{towrite_fpath}_top20.txt', 'w') as file:
        for word, count in top_words:
            file.write(f'{word}: {count}\n')         # the frequency of each word is not required
    print('Done')
```

# Question 1:

```python
cv_path = './scistatistics_data/CV'
ml_path = './scistatistics_data/ML'
nlp_path = './scistatistics_data/NLP'

cv_corpus = get_corpus(cv_path)
ml_corpus = get_corpus(ml_path)
nlp_corpus = get_corpus(nlp_path)

cv_top_20_words = get_frequent_words(cv_corpus, 20)
ml_top_20_words = get_frequent_words(ml_corpus, 20)
nlp_top_20_words = get_frequent_words(nlp_corpus, 20)
```

```python
def get_num_overlapping(list1, list2):
    return len(set(list1).intersection(set(list2)))
```
