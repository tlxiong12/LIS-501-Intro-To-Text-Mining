```python
def get_corpus(fd_path): 
    '''
       Input: 
            fd_path: a string of input folder path 
            
       Output:
            corpus_to_string: a string of content from all text files in the input folder
    '''
    
    data = []
    files = os.listdir(fd_path)
    
    # read the content of each text file
    for fname in files:
        if fname.endswith('.txt'):
            f_content = open(fd_path + '/' + fname, mode='r', encoding = 'utf-8').read()
            data.append(f_content)
    
    # aggreage the list of texts into a single string
    corpus_to_string = ' '.join(data)
    
    return corpus_to_string


def get_frequent_words(corpus, k):
    '''
       Input: 
            corpus: a string of corpus
            k: the number of top frequent words to be selected, here k should be 20
       Output:
            top_k_words: a list of top k most frequent words and their frequency
    '''

    ## To be added
    
    return top_k_words


def get_num_overlapping(list1, list2):
    '''
       Input: 
            list1: top frequent words in corpus A 
            list2: top frequent words in corpus B 
       Output:
            number_shared_words: total number of distinct words shared in corpus A and B
    '''

    ## To be added
    
    return number_shared_words
```

```python
cv_path = './scistatistics_data/CV'
ml_path = './scistatistics_data/ML'
nlp_path = './scistatistics_data/NLP'

cv_corpus = get_corpus(cv_path)
ml_corpus = get_corpus(ml_path)
nlp_corpus = get_corpus(nlp_path)
```

## Question 1:

* Get the top 20 most frequent words in each research field corpus.

```python
from collections import Counter
import spacy
import en_core_web_sm

nlp = en_core_web_sm.load()

text_cv = nlp(cv_corpus)
text_ml = nlp(ml_corpus)
text_nlp = nlp(nlp_corpus)

cv = [
    token.text.lower()
    for token in text_cv
    if token.is_alpha and not token.is_stop
]
ml = [
    token.text.lower()
    for token in text_ml
    if token.is_alpha and not token.is_stop
]

nlp = [
    token.text.lower()
    for token in text_nlp
    if token.is_alpha and not token.is_stop
]


top20_cv = Counter(cv).most_common(20)
top20_ml = Counter(ml).most_common(20)
top20_nlp = Counter(nlp).most_common(20)

top20_cv
```

## Question 2:

* Aggregate the corpus across three research fields into a single corpus.
* Get the top 20 most frequent words in the aggregated corpus.

```python
corpus_agg = cv_corpus + " " + ml_corpus + " " + nlp_corpus

nlp = spacy.load("en_core_web_sm")

text = nlp(corpus_agg)

words = [
    token.text.lower()
    for token in text
    if token.is_alpha and not token.is_stop
]

top20 = Counter(words).most_common(20)
top20
```
