## Python data structures

You'll use a lot of *lists, dictionaries, and sets*.

A *list* is an ordered sequence of elements, and can contain duplicate elements.

````python
sentence = ['A', 'list', 'is', 'an', 'ordered', 'sequence', 'of', 'elements', 'and', 
            'can', 'contain', 'duplicate', 'elements']
sentence
````

A *set* is not ordered, and cannot contain duplicates.

````python
sentence_set = set(sentence)
sentence_set
````

A *dictionary* maps *keys* to *values*.

````python
customers = {2086: 'Louis-Dreyfus, Julia',
            2122: 'Vaidya, Omi',
            2201: 'Atkinson, Rowan'
            }
````

````python
customers[2201]
````

A for loop *iterates* across a sequence that can be defined in several ways.

````python
for i in range(0, 4):
    print(i)
````

````python
for idx, word in enumerate(sentence):
    print(str(idx) + ' ' + word, end = ' ')
````

````python
for cust_id, cust_name in customers.items():
    print(cust_id, cust_name)
````

## From lists to numpy arrays


Python has lots of built-in abilities, but sometimes you need a data structure or function that isn't in the basic library. In that case you can "import" a whole module, or (as in the case below) a particular part of one.

````python
# typing the following command line in your created conda environment
# conda install conda-forge::numpy
# conda install matplotlib

import numpy as np
from matplotlib import pyplot as plt
````

Let's make a quadratic curve:

````python
x = []
y = []

for i in range(11):
    x.append(i)
    y.append(i * i)

plt.plot(x, y)
````

````python
x_n = np.arange(11)
y_n = x_n ** 2

plt.plot(x_n, y_n)
````
